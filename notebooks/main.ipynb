## Import packages

```
import torch
import torch.nn.functional as F
import torch.nn as nn
```

```
class Classifier(nn.Module):
  def __init__(self, in_nc, out_nc):
    super(Classifier, self).__init__()
    layers = [nn.Conv2d(in_nc, 16, kernel_size=4)]

    nc = 16
    for _ in range(5):
      new_nc = nc * 2
      layers += [nn.Conv2d(nc, new_nc, kernel_size=4)]
      nc = new_nc

    self.layers = nn.ModuleList(layers)

  def forward(self, x):
    for layer in self.layers:
      x = layer(x)
      x = F.relu(x)
```

```
classifier = Classifier(3, 3)

x = torch.ones(1, 3, 100, 100)

out = classifier(x)

```
